#### 1) Please finish this exercise that we started in class: One of the examples in chap07.ipynb compares two mixtures of complex sinusoids, one with offset=0 and one with offset=1.  The result is a different looking waveform, because "rotating" each component by the same phase offset does not shift each component by the same amount of time.  Can you construct a vector of complex amplitudes that shifts all components by the same amount of time?

```python
amps = numpy.array([0.6, 0.25, 0.1, 0.05])
freqs = [100, 200, 300, 400]
framerate = 11025
ts = numpy.linspace(0, 1, framerate)
i = complex(0, 1)

def synthesize2(amps , freqs, ts):
    args = numpy.outer(ts, freqs)
    M = numpy.exp(i * PI2 * args)
    ys = numpy.dot(M, amps)
    return ys

ys = synthesize2(amps, freqs, ts)

n=400
thinkplot.plot(ts[:n], ys.real[:n])
thinkplot.plot(ts[:n], ys.imag[:n])
thinkplot.config(ylim=[-1.15, 1.05])
```

The above plot is a short segment of the real and imaginary parts of a complex sinusoid. The sinusoid was synthesized from the amplitudes, frequencies, and times using linear algebra.

```python
amps2 = amps * numpy.exp(5j * numpy.array(freqs))
ys2 = synthesize2(amps2, freqs, ts)

n=400
thinkplot.plot(ts[:n], ys2.real[:n])
thinkplot.plot(ts[:n], ys2.imag[:n])
thinkplot.config(ylim=[-1.15, 1.05])
```

Taking the vector of amplitudes above and scaling it by a complex exponential applies a phase offset to the signal. However, while shifting the signal in time, it also changes the shape of the imaginary part of the signal. In order to shift the signal in time without changing the shape of the imaginary part of the signal, the vector of amplitudes needs to be multiplied by a complex exponential that applies a phase offset multiplied by the frequencies. This causes the phase offset to apply to the time domain of the signal and shifts the signal in time without applying the rotation of the phase offset.

#### 2) In Chapter 7 I represented the DFT as matrix multiplication, but the more common presentation is as a summation. Read the Wikipedia page on DFT.  Write an implementation of DFT based on this summation and compare your results to my implementation and/or numpy.fft.fft.

```python
signal = thinkdsp.SquareSignal(freq=440, amp=1)
wave = signal.make_wave(start=0, duration=1, framerate=11025)
wave.make_spectrum().plot()
```

The spectrum above is of a square wave with a frequency of 440 Hz. This spectrum has been calculated so far this semester using the DFT and will be used to compare the resulting spectrum of my implementation of the DFT.

```python
def dft(ys):
    N = len(ys)
    freqs = numpy.arange(N)
    total = 0
    for i in range(N-1):
        n = ys[i]  #takes a single element of the wave array
        total += n*numpy.exp((-1j * PI2 * freqs * i)/N)  #increments total by product of complex expnential and n
    return numpy.absolute(total)  #total contains complex numbers. numpy.absolute returns the amplitudes

amps = dft(wave.ys)
fs = numpy.arange(len(amps)) * wave.framerate/len(wave.ys) #arranges a vector of frequencies the amplitudes map to

thinkplot.plot(fs[:wave.framerate/2], amps[:wave.framerate/2])  #only plots 1/2 the spectrum. Plotting more shows aliasing.
```

The above spectrum is the result of my implementation of DFT. This algorithm takes a significant amount of time to run but produces a spectrum that is practically identical to the previous spectrum. However, since it takes a significant amount of time to run, the matrix multiplication implementation of DFT is superior.

#### 3) In Chapter 7, I showed how we can express the DFT and inverse DFT as matrix multiplications. These operations are relatively fast, taking time proportional to N2, where N is the length of the wave array. That would be fast enough for many applications, but it turns out that there is a faster algorithm, the Fast Fourier Transform (FFT), which takes time proportional to N logN.  Read about the FFT at http://en.wikipedia.org/wiki/Cooley-Tukey_FFT_algorithm, and write an implementation. Hint: I suggest you write a simple version as a recursive function; don’t worry about “data reordering, bit reversal, and in-place algorithms”.

```python
def fft(ys):
    N = len(ys)
    if N == 1:
        return ys
    else:
        even = fft(ys[::2])
        odd = fft(ys[1::2])

        res = [0] * N
        
        for m in xrange(N/2):
            res[m] = even[m] + numpy.exp((-1j * PI2 * m)/N) * odd[m]
            res[m+N/2] = even[m] - numpy.exp((-1j * PI2 * m)/N) * odd[m]
            
        return res

hs = fft(wave.ys[:2**13])
amps = numpy.absolute(hs)
fs = numpy.arange(len(amps)) * wave.framerate/len(amps)

thinkplot.plot(fs[:wave.framerate/2], amps[:wave.framerate/2])
```

In this implementation of FFT, which uses recursion, the input wave array must have a length that is a power of 2. This algorithm takes a wave and slices it down into the smallest segments that are divisible by 2 for both odd and even elements. The function recursively calculates the FFTs and pieces the results back together, using the distinction between the odd and even elements.

The spectrum above does not look exactly like the previous two spectrums, all of which are of a square wave at 440 Hz. The amplitudes in this spectrum are decreased and while it does still decay at a rate of 1/f at lower frequencies, at higher frequencies it appears to stop decaying.

#### 4) Using the timing code from PS03, compare the performance of my implementation of autocorr from Chapter 5, my fft_autocorr from Chapter 8, and numpy.correlate.  Can you tell whether NumPy is using the FFT algorithm?

```python
def corrcoef(xs, ys):
    return numpy.corrcoef(xs, ys, ddof=0)[0, 1]

def serial_corr(wave, lag=1):
    N = len(wave)
    y1 = wave.ys[lag:]
    y2 = wave.ys[:N-lag]
    corr = corrcoef(y1, y2)
    return corr

def autocorr(wave):
    """Computes and plots the autocorrelation function.

    wave: Wave
    """
    lags = range(len(wave.ys)//2)
    corrs = [serial_corr(wave, lag) for lag in lags]
    return lags, corrs
```

```python
def zero_pad(array, n):
    res = numpy.zeros(n)
    res[:len(array)] = array
    return res

def fft_convolve(signal, window):
    fft_signal = numpy.fft.fft(signal)
    fft_window = numpy.fft.fft(window)
    return numpy.fft.ifft(fft_signal * fft_window)

def fft_autocorr(signal):
    N = len(signal)
    window = signal[::-1]
    signal = zero_pad(signal, 2*N)
    window = zero_pad(window, 2*N)

    corrs = fft_convolve(signal, window)
    corrs = corrs[N//2: 3*N//2]
    return corrs
```

```python
ns = 2 ** numpy.arange(4, 12)
print(ns)
```

```python
results1 = []
results2 = []
results3 = []
for N in ns:
    signal = thinkdsp.UncorrelatedGaussianNoise()
    noise = signal.make_wave(duration=1.0)
    noise.ys = noise.ys[:N]
    result1 = %timeit -o autocorr(noise)
    result2 = %timeit -o fft_autocorr(noise.ys)
    result3 = %timeit -o numpy.correlate(noise.ys, noise.ys, 'full')
    results1.append(result1)
    results2.append(result2)
    results3.append(result3)

bests1 = [result.best for result in results1]
bests2 = [result.best for result in results2]
bests3 = [result.best for result in results3]
```

```python
thinkplot.preplot(3)
thinkplot.plot(ns, bests1, label='autocorr')
thinkplot.plot(ns, bests2, label='fft_autocorr')
thinkplot.plot(ns, bests3, label='numpy.correlate')
thinkplot.config(xscale='log', yscale='log', loc='upper right')
```

Looking at the plot above of the performance of all three algorithms, numpy.correlate is the fastest algorithm for less values but increasing complexity increases the run time exponentially, which then makes fft_autocorr the fastest algorithm. autocorr has a slope of 1, which corresponds to N complexity. fft_autocorr has a slope of 0.43934 and is not linear. This algorithm is Nlog(N) complexity. Finally, numpy.correlate has a slope of 1.23869 but also appears exponential. This probably means numpy.correlate has a complexity between N and $N^2$. Since numpy.correlate is not around a complexity of Nlog(N), it probably doesn't use the FFT algorithm.
