#### 6) A sawtooth signal has a waveform that ramps up linearly from -1 to 1, then drops to -1 and repeats. See   Write a class called SawtoothSignal that extends Signal and provides evaluate to evaluate a sawtooth signal.  Note: there is an implementation of SawtoothSignal in thinkdsp.py; you should avoid looking at it until you have written your own.

```python
class SawtoothSignal(thinkdsp.Sinusoid):
    """Represents a sawtooth signal."""
    
    def evaluate(self, ts):
        """Evaluates the signal at the given times.

        ts: float array of times
        
        returns: float wave array
        """
        cycles = self.freq * ts + self.offset / thinkdsp.PI2
        frac, _ = numpy.modf(cycles)
        ys = thinkdsp.normalize(thinkdsp.unbias(frac), self.amp)
        return ys

sawtooth_sig = SawtoothSignal(freq=440, amp=1, offset=0)
sawtooth_wave = sawtooth_sig.make_wave(duration=0.01, start=0, framerate=11025)
sawtooth_wave.plot()
```

#### 7) Compute the spectrum of a sawtooth wave. How does the harmonic structure compare to triangle and square waves?

```python
sawtooth_wave = sawtooth_sig.make_wave(duration=2, start=0, framerate=11025)
sawtooth_spectrum = sawtooth_wave.make_spectrum()
sawtooth_spectrum.plot()
```

The sawtooth wave has both even and odd harmonics and decays at a rate of 1/f in comparison to the triangle wave which only has even harmonics and decays at 1/f^2 and the square wave which only has odd harmonics and decays at a rate of 1/f. It would appear that the sawtooth wave is a combination of both the harmonics of the triangle and square waves; however it still decays at a rate of 1/f like a square wave.
