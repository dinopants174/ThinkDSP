#### 1) Following the example in chap07.ipynb, make a synthesis matrix (the IDFT matrix) using a different range of frequencies, and see what difference it makes.  The usual version of IDFT uses frequencies 0 through N-1.  What happens if you go from N to 2N-1?

```python
def dft(ys):
    N = len(ys)
    ts = numpy.arange(N) / float(N)
    freqs = numpy.arange(0, N)
    args = numpy.outer(ts, freqs)
    M = numpy.exp(1j * PI2 * args)
    hs = M.conj().transpose().dot(ys)
    return hs

def idft(amps):
    N = len(amps)
    ys = dft(amps) / N
    return ys

def dft2(ys):
    N = len(ys)
    ts = numpy.arange(N) / float(N)
    freqs = numpy.arange(N, 2*N)
    args = numpy.outer(ts, freqs)
    M = numpy.exp(1j * PI2 * args)
    hs = M.conj().transpose().dot(ys)
    return hs

def idft2(amps):
    N = len(amps)
    ys = dft2(amps) / N
    return ys
```

So the first dft function assembles the synthesis matrix using a range of frequencies from 0 to N-1. The IDFT matrix is the conjugate transpose of the matrix generated in the dft function. Changing the frequencies in the DFT matrix then also changes the frequencies in the IDFT matrix, since the two are related by computing the complex conjugate transpose. As a result, the dft2 function computes the synthesis matrix using frequencies ranging from N to 2N-1 and subseuqently, the IDFT synthesis matrix also contains this new range of frequencies.

```python
amps = abs(thinkdsp.TriangleSignal(freq=1).make_wave(duration=10, framerate=500).make_spectrum().hs)
ys1 = idft(amps)
thinkplot.plot(ys1)
```

This is the result of taking the amplitudes of a triangle signal with a frequency of 1 Hz running for 10 seconds and applying the inverse DFT to plot the wave. The IDFT synthesis matrix used for this plot contains frequencies ranging from 0 to N-1.

```python
ys2 = idft2(amps)
thinkplot.plot(ys2)
```

This is the result of taking the same amplitudes above and running them through the idft2 function. The IDFT synthesis matrix consists of frequencies from N to 2N-1. This plot is exactly the same as the previous plot, both of which have index in the ys vector as the x-axis. This means that changing the range of frequencies in the synthesis matrix or the IDFT matrix does not change the resulting wave output. This is because changing the range of frequencies changes the calculation of phase in the synthesis/IDFT matrix. It does not change the calculation of the amplitude. Since amplitudes are passed into the IDFT, it makes sense that changing the range of frequencies would not change the resulting wave, since the idft takes amplitudes, which have not changed, and not the phases which have changed, as input.

#### 2) In thinkdsp.py, the Spectrum methods low_pass and high_pass are ideal filters, in some sense: in the passband they are perfectly flat, and in the stopband they attenuate to 0 (or whatever value you want).  Invoking the principle that for every filter there is a corresponding window, see if you can figure out what window corresponds to these ideal filters.

```python
low_pass = []

for i in range(0, 40):
    if i < 20:
        low_pass.append(1)
    else:
        low_pass.append(0)

low_pass = numpy.array(low_pass)
freqs = numpy.linspace(0, 1, len(low_pass))
ts = numpy.linspace(0, 10, len(low_pass))

def synthesize2(amps, freqs, ts):
    args = numpy.outer(ts, freqs)
    M = numpy.exp(1j * PI2 * args)
    ys = numpy.dot(M, amps)
    return ys

ys = synthesize2(low_pass, freqs, ts)

thinkplot.plot(ts, ys)
```

So I simulated a low pass filter by creating a vector that contains 1s and then 0s. I used this vector as the amplitudes into the synthesize2 function, which given frequencies and times, computes the wave. I initially used the idft function given above or numpy.fft.ifft but I was dissatisfied with the resolution of the output. Using the synthesize2 function, it is clear that the corresponding window to a low-pass filter in the frequency domain is a sinc function in the time domain. In addition, looking at the peaks of the window, the window decays at a rate of 1/t.

```python
high_pass = []

for i in range(0, 40):
    if i < 20:
        high_pass.append(0)
    else:
        high_pass.append(1)


high_pass = numpy.array(high_pass)
freqs = numpy.linspace(0, 1, len(high_pass))
ts = numpy.linspace(0, 5, len(high_pass))
ys = synthesize2(high_pass, freqs, ts)
thinkplot.plot(ys)
```

I simulated a high pass filter in the same way as above, by creating a vector of 0s and then 1s. The window in the time domain that corresponds to a high pass filter in the frequency domain is still a sinc function that is more complex. However, same as the low pass filter's corresponding window, the peaks in this window also decay at a rate of 1/x. 

#### 3) So far we have focused attention on the effect filters have on the magnitude of each frequency component, but they also affect the phase offset.  For the boxcar and Gaussian windows, plot the phase shift as a function of frequency.

```python
signal = thinkdsp.SquareSignal(freq=440)
wave = signal.make_wave(duration=1, framerate=44100)
segment = wave.segment(duration=0.01)

hs = numpy.fft.fft(segment.ys)
phase_offset = numpy.angle(hs)
fs = numpy.arange(len(hs)) * wave.framerate/len(wave.ys)

thinkplot.plot(fs, phase_offset)
```

Above is plot of the phase offsets of a square signal at a frequency of 440 Hz as a function of frequency. The result appears to resemble a square wave where each period has been slanted by the same amount. This plot is for comparison to the effect on the phase offset of applying the boxcar and Gaussian windows.

```python
boxcar = numpy.ones(11)
boxcar /= sum(boxcar)

ys = numpy.convolve(segment.ys, boxcar, mode='valid')
hs = numpy.fft.fft(ys)
phase_offset = numpy.angle(hs)
fs = numpy.arange(len(hs)) * wave.framerate/len(wave.ys)

thinkplot.plot(fs, phase_offset)
```

If I manipulate my window that I am convolving with, I see that the resulting plot of phase offset vs. frequency essentially starts to begin looking like the previous plot. This indicates that applying the boxcar window actually does not change the phase offset as a function of frequency. For this example, my boxcar window needed to contain 11 values and decreasing the window just makes the resulting signal look more and more like the previous plot of the original signal. Increasing the size of the window makes the plot look less and less like the previous plot.

```python
gaussian = scipy.signal.gaussian(M=11, std=2)
gaussian /= sum(gaussian)

ys = numpy.convolve(segment.ys, gaussian, mode='valid')
hs = numpy.fft.fft(ys)
phase_offset = numpy.angle(hs)
fs = numpy.arange(len(hs)) * wave.framerate/len(wave.ys)

thinkplot.plot(fs, phase_offset)
```

In this plot, the Gaussian window was convolved with the original signal. The resulting plot of phase offset vs. frequency also looks similar to the original plot. This again indicates that convolving the original signal with the Gaussian window does not change the plot of phase offset as a function of frequency. Increasing the standard deviation makes the output plot look less and less like the original plot while decreasing the standard deviation makes the output plot practically indistinguishable from the original plot. This is due to changing the size of the window. However, we are still really seeing that convolution with the window does not change the phase offset.

#### 4) Taking the difference of successive elements in a wave array has the same effect as convolution with the window [1, -1].  The filter for this window is an increasing function of f, so you can think of diff as a high-pass filter.  One way to explain this behavior is that diff is a rough approximation of a derivative, and differentiation corresponds to a filter where amplitude is proportional to f.  If we use a better approximation of a derivative, we might expect the corresponding filter to be a better approximation of a line.  The window [-1/2, 2, -3/2] is a second order forward difference approximation of the first derivative.  Compute its DFT and see what the filter looks like.  Also try [1, -2, 1], which is approximates the second derivative.  What does the filter for the second derivative look like?  And how does that compare to the filter for this window?

```python
def zero_pad(array, n):
    res = numpy.zeros(n)
    res[:len(array)] = array
    return res

diff = zero_pad(numpy.array([1, -1]), 500)
diff = thinkdsp.Wave(diff, framerate=11025)
diff.make_spectrum().plot()
```

For this problem, I took the matrix [1, -1] that is an approximate definition of a derivative, and padded it with 500 0s. I then took that vector and set it as the vector of ys in a wave object. I then plotted the spectrum of this wave, which is the above plot. This plot looks like a high-pass filter, attenuating lower frequencies and allowing higher frequencies to pass. The above plot is an approximation of the derivative of a signal, which is also a high pass filter.

```python
better_diff = zero_pad(numpy.array([-0.5, 2, -1.5]), 500)
better_diff = thinkdsp.Wave(better_diff, framerate=11025)
better_diff.make_spectrum().plot()
```

The plot above is also an approximation of the derivative but is more accurate than the previous plot. This plot also resembles a high pass filter and is more accurate than the previous plot, since it attenuates the lower frequencies more and atteuates the higher frequencies less. This plot is more accurate because the matrix it uses is a second order forward approximation of the derivative.

```python
second_deriv = zero_pad(numpy.array([1, -2, 1]), 500)
second_deriv = thinkdsp.Wave(second_deriv, framerate=11025)
second_deriv.make_spectrum().plot()
```

The plot above is an approximation of the second derivative of a signal, which is also a high pass filter. However, this is really the derivative of high pass filter, since it is the second derivative of the signal. This is demonstrated by the increased curvature of the above plot, which again attenuates lower frequencies better and higher frequencies less.

This is a plot of the improved gunshot, which captures the loudest part of the signal, at the beginning of the gunshot. This loudest part is visible in the above plot and is audible in the sound wave below. This loudest part means won't overestimate the reverbs of the impulse response.

```python
saxophone.make_audio()
```

```python
spectrum = saxophone.make_spectrum()
```

The spectrum of the saxophone is calculated. Now, both the spectrum of the gunshot and the saxophone are calculated and we are in the frequency domain.
```python
output = (spectrum * transfer).make_wave()
output.normalize()
output.make_audio()
```
