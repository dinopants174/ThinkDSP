#### 2) Compute the autocorrelation of BitCoin prices. Does the autocorrelation function decay exponentially? Is there evidence of periodic behavior?

```python
def corrcoef(xs, ys):
    return numpy.corrcoef(xs, ys, ddof=0)[0, 1]

def serial_corr(wave, lag=1):
    N = len(wave)
    y1 = wave.ys[lag:]
    y2 = wave.ys[:N-lag]
    corr = corrcoef(y1, y2)
    return corr

def autocorr(wave):
    """Computes and plots the autocorrelation function.

    wave: Wave
    """
    lags = range(len(wave.ys)//2)
    corrs = [serial_corr(wave, lag) for lag in lags]
    return lags, corrs
```

```python
lags, corrs = autocorr(wave)
thinkplot.plot(lags, corrs)
```

The autocorrelation function of the bitcoin prices appears to be sinusoidal, as shown in the above plot. There is a peak in correlation of approximately 0.9 at a lag of 250. This peak would be indicative of periodic behavior of the price of bitcoin. Doing some simple math would allow us to estimate the fundamental frequency, solving for the period as lag / framerate, which is 250/1 days. The fundamental frequency then is 1/250 which is 0.004 cycles/day.

#### 4) Test the algorithmic complexity of analyze1, analyze2 and scipy.fftpack.dct.

```python
def analyze1(ys, freqs, ts):
    args = numpy.outer(ts, freqs)
    M = numpy.cos(PI2 * args)
    amps = numpy.linalg.solve(M, ys)
    return amps
```

```python
signal = thinkdsp.UncorrelatedGaussianNoise()
noise = signal.make_wave(duration=1.0)
ns = 2 ** numpy.arange(4, 12)
print(ns)
```

```python
results = []
for N in ns:
    ts = (0.5 + numpy.arange(N)) / N
    freqs = (0.5 + numpy.arange(N)) / 2
    ys = noise.ys[:N]
    result = %timeit -o analyze1(ys, freqs, ts)
    results.append(result)
```

```python
bests = [result.best for result in results]
thinkplot.plot(ns, bests)
thinkplot.config(xscale='log', yscale='log')
```

The plot above demonstrates that as complexity increases, the amount of time required the algorithm takes increases almost exponentially. The relationship is mostly linear but increasing complexity increases runtime to milliseconds.

```python
def analyze2(ys, freqs, ts):
    args = numpy.outer(ts, freqs)
    M = numpy.cos(PI2 * args)
    amps = M.dot(ys) / 2
    return amps
```

```python
results = []
for N in ns:
    ts = (0.5 + numpy.arange(N)) / N
    freqs = (0.5 + numpy.arange(N)) / 2
    ys = noise.ys[:N]
    result = %timeit -o analyze2(ys, freqs, ts)
    results.append(result)
```

```python
bests = [result.best for result in results]
thinkplot.plot(ns, bests)
thinkplot.config(xscale='log', yscale='log')
```

The analyze2 function, which uses orthogonal matrices to solve for the amplitudes, is less exponential as complexity increases and more of a linear function. In addition, as complexity increases, the runtime doesn't increase as much as the runtime for analyze1 did. When I ran it inititally, analyze2 was approximately 5 times faster than analyze1.

```python
results = []
for N in ns:
    ts = (0.5 + numpy.arange(N)) / N
    freqs = (0.5 + numpy.arange(N)) / 2
    ys = noise.ys[:N]
    result = %timeit -o scipy.fftpack.dct(ys)
    results.append(result)
```

```python
bests = [result.best for result in results]
thinkplot.plot(ns, bests)
thinkplot.config(xscale='log', yscale='log')
```

Compared to analyze1 and analyze2, scipy.fftpack.dct performs the best. Increasing complexity produces very little change in runtime and even at higher complexities, scipy.fftpack.dct produces a runtime on the order of microseconds, which is orders of magnitude better than the runtimes of milliseconds that analyze1 and analyze2 were producing.

#### 5) One of the major applications of the DCT is compression for both sound and images. In its simplest form, DCT-based compression works like this:  Break a long signal into segments.  Compute the DCT of each segment.  Identify frequency components with amplitudes so low they are inaudible, and remove them. Store only the frequencies and amplitudes that remain.  To play back the signal, load the frequencies and amplitudes for each segment and apply the inverse DCT.  Implement a version of this algorithm and apply it to a recording of music or speech.
```python
wave = thinkdsp.read_wave('100475__iluppai__saxophone-weep.wav')
wave = wave.segment(start=2, duration=5)
wave.make_spectrum().plot()
```

```python
all_amps = []
for i in range(5):
    test = wave.segment(start=wave.duration/5.0 * i, duration=wave.duration/5.0)
    test.ys = test.ys.tolist()
    test.ys = [float(i) for i in test.ys]
    amps = scipy.fftpack.dct(test.ys)
    all_amps.append(amps)
```

```python
final = thinkdsp.Wave(ys=numpy.zeros(len(amps)), framerate=len(amps))
for i in range(5):
    amps = abs(all_amps[i])
    for j in range(0, len(amps)):
        if amps[j] < 0.5*numpy.mean(amps):
            amps[j] = 0
    ys = scipy.fftpack.idct(amps)
    wave = thinkdsp.Wave(ys=ys, framerate=len(amps))
    final = final | wave
```
